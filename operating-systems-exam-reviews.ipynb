### CPU Scheduling [Lecutre 9-10]
        Assigning process to the CPU, via short term scheduling
        
        1. Nature of process
            * CPU burst / I/O processes
        
        2. When are scheduling decisions made?
            a. A new process arrives
            b. A process blocked (Another process must be served)
            c. A process terminates (Another process must be served)
            d. A process goes from blocked to ready
            e. A process goes from running to ready
       

           non-preemptive = a process runs until completion or it blocks
           preemptive = running process is interrupted and placed in the ready queue

           How to evaluate different schedulers
           1. CPU Utilization
           2. Turn-around time
           3. Response time
           4. Throughput
           5. Waiting Time
           6. Fairness / priority/ deadline
           
###### First Come First Serve (Non-preemptive)
        
        Gives CPU to the first requesting process
        
        adv: Simple
        disadv: Average waiting time
        
        
###### Shortest Job First (Non-preemptive)
        Associate with every process the length of its next CPU burst
        
        Essnetially assign the CPU to the process with the shortest CPU burst -> STF is optimal minimum average 
        waiting time
        
        disadv: Shortest job first may lead to process starvation

###### Shortest Remaining Time First (Preemptive)
        Shorter jobs being given immedieate attention over all via context switching
        
        adv: Lower turn-around time than STF
       

###### Highest Response Ratio Next (Non-preemptive)
        Minimize the normalized turnaround time
        
        R = (w + s) / s
        
        with s being the expected serivce time
        with r being the response ratio
        with w being the time spent waiting
        
        Which ever process has the highest R that is the process to be be picked


###### Round Robin (Preemptive)
        Using a clock to base everything off of, in this case by using a quantum or time slice thats how its able 
        to manage.
        
        Interrupts are guarnteed every time slice
         * 80% if the CPU burst to be shorter than q.
         * maximum waiting time = (n-1)q
         
            with n being process
            with q being quantum
            
        Short Quantum => Overhead Context Switching
        Long Quantum => Approaches FCFS
        
###### Priority Scheduling  (Preemptive / non-preemptive)
        Select process with highest priority
            * This could be internal or external
        Starvation can happen
        Aging -> Effectively increases priorty over time as processes have to wait.
        
###### Multi-level Queue Scheduling 
        Classify processes into different groups
        -> Each group is assigned a ready queuue
