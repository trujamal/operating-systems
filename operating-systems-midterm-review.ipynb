### What is an OS? [Lecture 1]
      A program that manges the hardware for the application and system programs. Task between the application and 
      hardwware. operating system is found to be defining program.
      
      Ease of use is the main goal of any operating system

          - Systems (Resource Utilization)

###### Operation
      When a program in firmware exectues, it intializes aspects of the system, and loads the OS into emmory, the 
      OS will execute the first process and waits for the event to happen (Essentially in a while loop)
      
      Events will typically be signaled by an interrupt. [Hardware and software]
      
      Why Interrupts?
      
      * To keep the CPU as busy as possible
      * Also to improve CPU Utilization due to I/O devices being much slower than the CPU
      
      Polling -> Is the act of checking every second if a process has completed yet, this is where interrupts come 
      into play.
      When interpreting the interrupt processing, PC, PSW, and Registered all need to be stored.
      ISR = Interrupt Service Routine
      
### API's [Lecutre 2]

      Why API's?
      
      Advantages include:
         * Portable
         * Simple | Safe
         * Elegant
         * Version Control
      
      * API's provide a set of functions (Parameters + Return Values) that are available to the programmer.
      * System Calls are involved within the API.
      
      There are 7 types of syscalls, they go as follows:
      1. Proces Management (execute, terminate, fork, kill)
      2. File Manupulation (open, delete, read, close, create, write)
      3. Device Manipulation (read, power, unmount, write, mount)
      4. Information Maintaince (get, set, time | date)
      5. Communication (send, recieve, read, write, socket)
      6. Protecion (chmod, user group ids)
      7. Memory Management (malloc, mmap and munmap)
      
      Most of these interrupt services have some form of overlap within them.
      
###### OS Strucutre
       Two different approaches, simple vs layered approach
       
       1. Simple
          Examples include MS DOS
          
          General design is a monolithic kernal, which acts a whole system, when one thing goes wrong the whole 
          strucutre falls apart. OS interfaces through hardware -> kernal -> syscalls

       2. Layered Approach
           Modular System Design
           UI -> Outtermost layer
           HW -> Inner Most
           
           Advantages include:
              * Easier to debug
              * Every layer contains abstraction
              * Hidding infomration
              * Ability to debug each layer independently
           
           Disadvantages: 
              * Dependencies are not clear cut
              * Determining functionality to a particular layer
              * Definining the amount of layers
              * Less Efficent
              
         3. Micro Kernels
           Small kernal that is well debugged, that includes interprocess communication
           
           Adv (Component Based):
           * Debugging
           * High fault tolearance
           * Easy to Extend
           * A failing service does not crash the main kernal
           
           Dis:
           * Alot of message passing
           * High overhead due to message passing.
           
         4. Modules
             Dynamic  loadable moduels

### Process Management [Lecutre 4]
       A Program in execution, that goes through mulitple states:
       * New: Process being created
       * Running: Using the CPU, executing instructions
       * Blocked: Blocked waiting for idle (waiting for an interupt)
       * Terminated: Process is exiting
       * Ready: Waiting for the CPU 
       
       In terms of the actual diagram itself Running -> Blocked then blocked to Ready
       Terminated -> exit();
       
       PCB (Proces Control Block) -> How the system views processes
       
       Long-term Scheduler vs Short-term Schedueler vs Medium-term
       L: Selects process memory
       S: Select a proces to execute on the CPU
       M: Swap processes in/out of memory to/ from disk
       
       - CPU-Bound: Process spend most of its time executing
       - I/O Bound: Process spends most of its time doing I/O's
       
       Need the right mix of info/data
